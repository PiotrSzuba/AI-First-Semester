```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1a
erdos_renyi = nx.erdos_renyi_graph(100, 0.05)
```

\

```python
#3a
data_frame = pd.read_table("radoslaw_email_email", names=["source", "target", "space", "weight", "timestamp"], sep=" ")
data_frame = data_frame.drop(["space", "weight", "timestamp"], axis=1)
data_frame.head(20)
```

```python
#3a
radoslaw_email = nx.from_pandas_edgelist(data_frame, source="source", target="target")

nx.draw(radoslaw_email, node_size=10)
```

```python
nodes = radoslaw_email.nodes()

print("Nodes:", len(nodes))
```

```python
edges = radoslaw_email.edges()

print("Edges:", len(edges))
```

```python
radoslaw_email.remove_edges_from(nx.selfloop_edges(radoslaw_email))
```

```python
nodes = radoslaw_email.nodes()

print("Nodes:", len(nodes))
```

```python
edges = radoslaw_email.edges()

print("Edges:", len(edges))
```

Po usuniciu ptli liczba krawdzi zmnieszya si o jeden.\
Liczba wierzchokw nie zmienia si

```python
#3b
degrees = dict(radoslaw_email.degree())
print(degrees)

degrees_values = list(degrees.values())
plt.hist(degrees_values, bins='auto')
plt.title("Histogram stopni wzw")
plt.xlabel("Stopie wza")
plt.ylabel("Liczba wzw")
plt.show()

betweenness = nx.betweenness_centrality(radoslaw_email)
print(betweenness)

betweenness_values = list(betweenness.values())
plt.hist(betweenness_values, bins='auto')
plt.title("Histogram miary porednictwa (betweenness)")
plt.xlabel("Warto miary porednictwa")
plt.ylabel("Liczba wzw")
plt.show()

closeness = nx.closeness_centrality(radoslaw_email)
print(closeness)

closeness_values = list(closeness.values())
plt.hist(closeness_values, bins='auto')
plt.title("Histogram miary bliskoci (closeness)")
plt.xlabel("Warto miary bliskoci")
plt.ylabel("Liczba wzw")
plt.show()
```

```python
from scipy.stats import kendalltau

betweenness_rank = sorted(betweenness.items(), key=lambda x: x[1], reverse=True)
closeness_rank = sorted(closeness.items(), key=lambda x: x[1], reverse=True)
degree_rank = sorted(degrees.items(), key=lambda x: x[1], reverse=True)

betweenness_rank = [value for _, value in betweenness_rank]
closeness_rank = [value for _, value in closeness_rank]
degree_rank = [value for _, value in degree_rank]

print("Kendall tau midzy betweenness a closeness:", kendalltau(betweenness_rank, closeness_rank)[0])
print("Kendall tau midzy betweenness a degree:", kendalltau(betweenness_rank, degree_rank)[0])
print("Kendall tau midzy closeness a degree:", kendalltau(closeness_rank, degree_rank)[0])
```

warto 1 oznacza pen zgodno rankingw\
warto -1 oznacza pen niezgodno rankingw\
warto 0 oznacza brak zwizku midzy rankingami\
\
W tym przypadku rankingi s bliskie wartoci 1 tak wic te rankingi s zgodne ze sob

```python
#3d
n_components = nx.number_connected_components(radoslaw_email)

print(n_components)
```

Przy takiej reprezentacji danych mog by wierzchoki bez krawdzi. W grafie reprezentowanym przez krawdzie w postaci DataFrame\
wierzchoki s definiowane jako unikalne wartoci w kolumnach "source" i "target".\
Jeli jaki wierzchoek nie ma krawdzi, to oznacza to tylko, e wiersza z takim wierzchokiem nie ma w DataFrame'ie.

```python
#3e
avg_path = nx.average_shortest_path_length(radoslaw_email)
diameter = nx.diameter(radoslaw_email)

print("rednica grafu:", diameter)
print("rednia dugo cieki:", avg_path)
print("Jaki procent wartoci rednicy grafu to warto avg_path", 100 * avg_path/diameter, "%")
```

```python
#3f
sent_count = data_frame.groupby(['source', 'target']).size()
all_sent_count = data_frame.groupby('source').size()

for index, row in data_frame.iterrows():
    source = row['source']
    target = row['target']
    weight = sent_count[source][target] / all_sent_count[source]
    data_frame.at[index, 'weight'] = weight

data_frame.head()
di_radoslaw_email = nx.from_pandas_edgelist(data_frame, source="source", target="target", create_using=nx.DiGraph)
```

```python
nx.draw(di_radoslaw_email, node_size=10)
```
