```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1a
erdos_renyi = nx.erdos_renyi_graph(100, 0.05)
```

```python
#1b
nodes = erdos_renyi.nodes()
edges = erdos_renyi.edges()

print(nodes)
print(edges)
```

```python
#1c
degree = erdos_renyi.degree()

print(degree)
```

```python
#1d
largest_cc = list(max(nx.connected_components(erdos_renyi), key=len))
number_of_cc = len(largest_cc)
print(number_of_cc)
```

```python
#1e
pr = nx.pagerank(erdos_renyi)
nx.draw(erdos_renyi, node_size=[v*2000 for v in pr.values()], cmap=plt.cm.Blues)
```

```python
#2a
barabasi_albert = nx.barabasi_albert_graph(1000, 3)
```

```python
#2b
nx.draw_kamada_kawai(barabasi_albert, node_size=6)
```

```python
#2c
closeness_values = nx.closeness_centrality(barabasi_albert)

most_central_node = max(closeness_values, key=closeness_values.get)

print("Node number:", most_central_node)
```

```python
#2d
diameter = nx.diameter(barabasi_albert)

print(diameter)
```

Różnice pomiędzy grafem Barabási-Albert i Erdős-Rényi:\
Barabási-Albert charakteryzuje się zjawiskiem przyklejanai wierzchołków oznacza to że wierzchołki które już mają dużo krawędzi mają większe szanse na utworzenie nowych.\
Erdős-Rényi stopień wierzchołków jest określany losowo (nie ma preferencji dla wierzochłko o wysokim stopniu)\
\
Barabási-Albert rozkład stopni wierzchołków potęgowy co oznacza że istnieją w nim wierzchołki o bardzo dużym stopniu ale większość wierzchołków ma niski stopień\
Erdős-Rényi tutaj rozkład stopni wierzchołków jest bardziej jednostajny\
\
Graf Barabási-Albert jest bardziej skłonny do tworzenia spójnych składowych niż graf Erdős-Rényi, ponieważ preferencyjne przyłączanie wierzchołków prowadzi do powstawania klastrów\
i bardziej rozbudowanej struktury siec\

```python
#3a
data_frame = pd.read_table("radoslaw_email_email", names=["source", "target", "space", "weight", "timestamp"], sep=" ")
data_frame = data_frame.drop(["space", "weight", "timestamp"], axis=1)
data_frame.head(20)
```

```python
#3a
radoslaw_email = nx.from_pandas_edgelist(data_frame, source="source", target="target")

nx.draw(radoslaw_email, node_size=10)
```

```python
nodes = radoslaw_email.nodes()

print("Nodes:", len(nodes))
```

```python
edges = radoslaw_email.edges()

print("Edges:", len(edges))
```

```python
radoslaw_email.remove_edges_from(nx.selfloop_edges(radoslaw_email))
```

```python
nodes = radoslaw_email.nodes()

print("Nodes:", len(nodes))
```

```python
edges = radoslaw_email.edges()

print("Edges:", len(edges))
```

Po usunięciu pętli liczba krawędzi zmnieszyła się o jeden.\
Liczba wierzchołków nie zmieniła się

```python
#3b
degrees = dict(radoslaw_email.degree())
print(degrees)

degrees_values = list(degrees.values())
plt.hist(degrees_values, bins='auto')
plt.title("Histogram stopni węzłów")
plt.xlabel("Stopień węzła")
plt.ylabel("Liczba węzłów")
plt.show()

betweenness = nx.betweenness_centrality(radoslaw_email)
print(betweenness)

betweenness_values = list(betweenness.values())
plt.hist(betweenness_values, bins='auto')
plt.title("Histogram miary pośrednictwa (betweenness)")
plt.xlabel("Wartość miary pośrednictwa")
plt.ylabel("Liczba węzłów")
plt.show()

closeness = nx.closeness_centrality(radoslaw_email)
print(closeness)

closeness_values = list(closeness.values())
plt.hist(closeness_values, bins='auto')
plt.title("Histogram miary bliskości (closeness)")
plt.xlabel("Wartość miary bliskości")
plt.ylabel("Liczba węzłów")
plt.show()
```

```python
from scipy.stats import kendalltau

betweenness_rank = sorted(betweenness.items(), key=lambda x: x[1], reverse=True)
closeness_rank = sorted(closeness.items(), key=lambda x: x[1], reverse=True)
degree_rank = sorted(degrees.items(), key=lambda x: x[1], reverse=True)

betweenness_rank = [value for _, value in betweenness_rank]
closeness_rank = [value for _, value in closeness_rank]
degree_rank = [value for _, value in degree_rank]

print("Kendall tau między betweenness a closeness:", kendalltau(betweenness_rank, closeness_rank)[0])
print("Kendall tau między betweenness a degree:", kendalltau(betweenness_rank, degree_rank)[0])
print("Kendall tau między closeness a degree:", kendalltau(closeness_rank, degree_rank)[0])
```

wartość 1 oznacza pełną zgodność rankingów\
wartość -1 oznacza pełną niezgodność rankingów\
wartość 0 oznacza brak związku między rankingami\
\
W tym przypadku rankingi są bliskie wartości 1 tak więc te rankingi są zgodne ze sobą

```python
#3d
n_components = nx.number_connected_components(radoslaw_email)

print(n_components)
```

Przy takiej reprezentacji danych mogą być wierzchołki bez krawędzi. W grafie reprezentowanym przez krawędzie w postaci DataFrame\
wierzchołki są definiowane jako unikalne wartości w kolumnach "source" i "target".\
Jeśli jakiś wierzchołek nie ma krawędzi, to oznacza to tylko, że wiersza z takim wierzchołkiem nie ma w DataFrame'ie.

```python
#3e
avg_path = nx.average_shortest_path_length(radoslaw_email)
diameter = nx.diameter(radoslaw_email)

print("Średnica grafu:", diameter)
print("Średnia długość ścieżki:", avg_path)
print("Jaki procent wartości średnicy grafu to wartość avg_path", 100 * avg_path/diameter, "%")
```

```python
#3f
sent_count = data_frame.groupby(['source', 'target']).size()
all_sent_count = data_frame.groupby('source').size()

for index, row in data_frame.iterrows():
    source = row['source']
    target = row['target']
    weight = sent_count[source][target] / all_sent_count[source]
    data_frame.at[index, 'weight'] = weight

data_frame.head()
di_radoslaw_email = nx.from_pandas_edgelist(data_frame, source="source", target="target", create_using=nx.DiGraph)
```

```python
nx.draw(di_radoslaw_email, node_size=10)
```
